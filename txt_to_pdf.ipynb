```python
import PyPDF2
```

```python
def extract_txt(file):
    with open(file,'rb') as f:
        r=PyPDF2.PdfReader(file)
        str=''
        for page in r.pages:
            str += page.extract_text()
    return str        
```

```python
result=extract_txt("pdf_sample.pdf")
print(result)
```
